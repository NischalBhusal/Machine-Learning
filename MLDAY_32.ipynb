```python


import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def calculate_metrics(y_true, y_pred, n_features=None):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    if n_features is not None:
        n_samples = len(y_true)
        adj_r2 = 1 - ((1 - r2) * (n_samples - 1)) / (n_samples - n_features - 1)
        return {"MSE": mse, "MAE": mae, "RMSE": rmse, "R2 Score": r2, "Adjusted R2 Score": adj_r2}
    else:
        return {"MSE": mse, "MAE": mae, "RMSE": rmse, "R2 Score": r2}

```

```python
y_true = np.array([3, -0.5, 2, 7, 4])
y_pred = np.array([2.5, 0.0, 2, 8, 3.5])

n_features = 1
metrics = calculate_metrics(y_true, y_pred, n_features=n_features)
print("Regression Metrics:")
for metric, value in metrics.items():
    print(f"  {metric}: {value:.4f}")
```
