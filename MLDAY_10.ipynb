```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import Binarizer

# Create a sample DataFrame
data = {
    'Score': np.random.randint(30, 100, 20),
    'Age': np.random.randint(18, 70, 20),
    'Income': np.random.randint(20000, 100000, 20)
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df.head())
```

```python
# Binning 'Score' into 4 equal-width bins
df['Score_Binned_Equal_Width'] = pd.cut(df['Score'], bins=4, labels=['Poor', 'Fair', 'Good', 'Excellent'])

# Binning 'Age' into 3 equal-frequency (quantiles) bins
df['Age_Binned_Equal_Frequency'] = pd.qcut(df['Age'], q=3, labels=['Young', 'Middle-aged', 'Senior'])

print("DataFrame after Binning:")
display(df[['Score', 'Score_Binned_Equal_Width', 'Age', 'Age_Binned_Equal_Frequency']].head())
```

```python
# Binarization using pandas.cut for 'Income' (e.g., above/below 50000)
df['Income_Binarized_Pandas'] = pd.cut(df['Income'], bins=[0, 50000, np.inf], labels=[0, 1], right=False)

# Binarization using sklearn.preprocessing.Binarizer for 'Score'
# Let's binarize 'Score' where anything >= 70 is 'high' (1), else 'low' (0)
binarizer = Binarizer(threshold=70)
df['Score_Binarized_Sklearn'] = binarizer.fit_transform(df[['Score']])

print("DataFrame after Binarization:")
display(df[['Income', 'Income_Binarized_Pandas', 'Score', 'Score_Binarized_Sklearn']].head())
```
