### Scraping Multiple Pages (Pagination)

Many websites paginate their content, meaning data is spread across several pages. To scrape all data, we need to:
1.  Identify the URL pattern for subsequent pages (e.g., `page/2/`, `page/3/`).
2.  Loop through pages, often by looking for a 'Next' button or link.
3.  Collect data from each page until no more 'Next' links are found.

For `http://quotes.toscrape.com/`, the pages are structured like `http://quotes.toscrape.com/page/1/`, `http://quotes.toscrape.com/page/2/`, and so on.

```python
import requests
from bs4 import BeautifulSoup

base_url = 'http://quotes.toscrape.com/'

all_quotes = []

page_number = 1

while True:
    if page_number == 1:
        current_url = base_url
    else:
        current_url = f"{base_url}page/{page_number}/"

    print(f"Scraping page: {current_url}")

    response = requests.get(current_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        quotes_on_page = soup.find_all('div', class_='quote')

        if not quotes_on_page:
            print("No more quotes found, ending scraping.")
            break

        for quote_div in quotes_on_page:
            text = quote_div.find('span', class_='text').text
            author = quote_div.find('small', class_='author').text
            all_quotes.append({'text': text, 'author': author})

        next_button = soup.find('li', class_='next')
        if next_button:
            page_number += 1
        else:
            print("Reached the last page, ending scraping.")
            break

    else:
        print(f"Failed to fetch content from {current_url}. Status code: {response.status_code}")
        break

print(f"\nTotal quotes scraped: {len(all_quotes)}")
for i, quote_data in enumerate(all_quotes):
    print(f"--- Quote {i+1} ---")
    print(f"Text: {quote_data['text']}")
    print(f"Author: {quote_data['author']}")
    print("\n")
```
